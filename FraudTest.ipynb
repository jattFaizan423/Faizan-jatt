```python
def map_gender(value):
    if str(value).strip().upper() == 'M':
        return 'Male'
    elif str(value).strip().upper() == 'F':
        return 'Female'
    else:
        return 'Unknown'
df['gender'] = df['gender'].apply(map_gender)
```

```python
df['gender'] = df['gender'].astype('category')
```

```python
df['Building_Number'] = df['Building_Number'].astype('int32')
```

```python
import re
def clean_street(address):
    address = str(address).strip()
    address = re.sub(r'\s+', ' ', address)

    match = re.match(r'^(\d+)\s+(.*)$', address)
    if not match:
        return address
    house_number = match.group(1)
    rest = match.group(2)

    apt_words = ['apartment', 'apt', 'suite', 'ste', 'unit', 'flat', 'floor', 'room', '#']
    apt_index = -1
    for word in apt_words:
        idx = rest.lower().find(word)
        if idx != -1:
            apt_index = idx
            break
    if apt_index != -1:
        street_name = rest[:apt_index].strip()
        apt_part = rest[apt_index:].strip()
        return f"{house_number}, {street_name}, {apt_part}"
    else:
        return f"{house_number}, {rest}"
df['street'] = df['street'].apply(clean_street)
```

```python
def convert_state_abbr(abbr):
    state_map = {
        'TX': 'Texas', 'NY': 'New York', 'PA': 'Pennsylvania', 'CA': 'California',
        'OH': 'Ohio', 'MI': 'Michigan', 'FL': 'Florida', 'IL': 'Illinois',
        'AL': 'Alabama', 'MO': 'Missouri', 'AR': 'Arkansas', 'MN': 'Minnesota',
        'WI': 'Wisconsin', 'VA': 'Virginia', 'SC': 'South Carolina',
        'NC': 'North Carolina', 'IA': 'Iowa', 'IN': 'Indiana', 'KY': 'Kentucky',
        'MD': 'Maryland', 'GA': 'Georgia', 'OK': 'Oklahoma', 'NJ': 'New Jersey',
        'WV': 'West Virginia', 'NE': 'Nebraska', 'MS': 'Mississippi',
        'KS': 'Kansas', 'LA': 'Louisiana', 'WY': 'Wyoming', 'TN': 'Tennessee',
        'OR': 'Oregon', 'WA': 'Washington', 'NM': 'New Mexico', 'ME': 'Maine',
        'CO': 'Colorado', 'SD': 'South Dakota', 'ND': 'North Dakota',
        'MA': 'Massachusetts', 'VT': 'Vermont', 'MT': 'Montana', 'AZ': 'Arizona',
        'UT': 'Utah', 'NH': 'New Hampshire', 'CT': 'Connecticut', 'ID': 'Idaho',
        'NV': 'Nevada', 'DC': 'District of Columbia', 'HI': 'Hawaii',
        'AK': 'Alaska', 'RI': 'Rhode Island'
    }
    return state_map.get(abbr, abbr)
```

```python
df['state'] = df['state'].apply(convert_state_abbr).astype('category')
```
