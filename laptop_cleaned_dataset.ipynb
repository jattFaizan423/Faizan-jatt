```python
import pandas as pd
import re

def format_cpu(text):
    # Flexible regex to match optional Intel, model, optional frequency
    match = re.search(r"(Intel)?\s*(.*?)(?:\s*@?\s*(\d+\.\d+GHz))?$", text.strip())

    if match:
        brand = match.group(1) or "Unknown"
        model = match.group(2).strip() or "Unknown"
        freq = match.group(3) or "Unknown"
        return f"{brand}, {model}, {freq}"

    # fallback in case regex fails completely
    return "Unknown, Unknown, Unknown"
```

```python
cpu_name = df['Cpu'].apply(format_cpu).str.strip().str.split(',').str[1]
```

```python
df.insert(9,'cpu_name',cpu_name)
```

```python
def Rom(value):
  if str(value) == '1':
    return '1024'
  elif str(value) == '2':
    return '2048'
  else:
    return value
```
