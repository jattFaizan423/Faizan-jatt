```python
def camera_extractor(text):

  if 'Quad' in text:
    return '4'
  elif 'Triple' in text:
    return '3'
  elif 'Dual' in text:
    return '2'
  elif 'No Data' in text:
    return 'No Data'
  else:
    return '1'
```

```python
num_rear_cameras = df['camera'].str.strip().str.split('&').str.get(0).apply(camera_extractor)
```

```python
num_front_cameras = df['camera'].str.strip().str.split('&').str.get(1).str.strip().fillna('No Data').apply(camera_extractor)
```
