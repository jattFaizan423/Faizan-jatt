```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.read_csv('/content/Airbnb_Open_Data.csv')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = df1.copy()
```

```python
df1.head(5)
```

```python
df['id'] = df['id'].astype(int)
```

```python
df['host id'] = df['host id'].astype(int)
```

```python
df['host_identity_verified'] = df['host_identity_verified'].fillna('unconfirmed').str.capitalize().astype('category')
```

```python
df['host name'] = df['host name'].fillna('Michael').str.strip().str.capitalize().astype('category')
```

```python
df['neighbourhood group'] = df['neighbourhood group'].fillna('Manhattan').str.strip().str.capitalize().astype('category')
```

```python
df['neighbourhood'] = df['neighbourhood'].fillna('Bedford-Stuyvesant').str.strip().str.replace('-',' ').str.replace(',',' ').str.capitalize().astype('category')
```

```python
df.rename(columns={'lat':'latitude','long':'longitude'},inplace=True)
```

```python
df['latitude'] = df['latitude'].fillna('40.72').astype(float).round(2)
```

```python
df['longitude'] = df['longitude'].fillna('-73.95').astype(float).round(2)
```

```python
df['country'] = df['country'].fillna('United States').str.strip().str.capitalize().astype('category')
```

```python
df['country code'] = df['country code'].fillna('US').str.strip().str.capitalize().astype('category')
```

```python
df['instant_bookable'].fillna(df['instant_bookable'].mode()[0], inplace=True)
df['instant_bookable'] = df['instant_bookable'].astype('bool')
```

```python
df['cancellation_policy'] = df['cancellation_policy'].fillna('flexible').str.strip().str.capitalize().astype('category')
```

```python
df['room type'] = df['room type'].fillna('Entire home').str.strip().str.split('/').str[0].astype('category')
```

```python
df['Construction year'] = df['Construction year'].fillna('2011.0').astype(float).astype('int32')
```

```python
df['price'] = df['price'].fillna('$67').str.strip().str.replace('$','').str.replace(',','').astype('int32')
```

```python
df['service fee'] = df['service fee'].fillna('$203').str.strip().str.replace('$','').astype('int32')
```

```python
df['minimum nights'] = df['minimum nights'].fillna('2.0').astype(float).astype('int32')
```

```python
df['number of reviews'] = df['number of reviews'].fillna('0.0').astype(float).astype('int32')
```

```python
df['last review'] = df['last review'].fillna('2019-06-23')
```

```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
df['reviews per month'] = df['reviews per month'].fillna('0.02').astype(float)
```

```python
df['review rate number'] = df['review rate number'].fillna('5.0').astype(float)
```

```python
df['calculated host listings count'] = df['calculated host listings count'].fillna('1.0').astype(float).astype('int32')
```

```python
df['calculated host listings count'] = df['calculated host listings count'].astype('int32')
```

```python
df['availability 365'] = df['availability 365'].fillna('197.0').astype(float).astype('int32')
```

```python
df['availability 365'] = df['availability 365'].astype('int32')
```

```python
df.rename(columns={
    'calculated host listings count': 'host_listings_count',
    'availability 365': 'days_available'
}, inplace=True)
```

```python
df.drop(['NAME','license','house_rules'],axis=1,inplace=True)
```
